```python
# Dependencies
import pymongo
import pandas as pd
import json
from flask import Flask, jsonify, render_template, Response
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.rescue_angels_db
dog_breeds_collection = db.dog_breeds
```

```python
# Read the csv file
url = 'breeds_v2_clean.csv'
dog_breeds_df = pd.read_csv(url)
dog_breeds_df = dog_breeds_df.fillna('')
dog_breeds_df.head()
```

```python
dog_breeds_df.dtypes
```

```python
# add dog breed to mondo db
for jdict in dog_breeds_df.to_dict(orient='records'):
    print(jdict)
    dog_breeds_collection.insert_one(jdict)
    
```
