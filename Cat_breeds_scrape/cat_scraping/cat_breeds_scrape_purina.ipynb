```python
cat_purina_list_of_dicts = []

for p in range(4):
    #####GO TO NEW BREED PAGE#####
    url = f"https://www.purina.com/cats/cat-breeds?page={p}"
    browser.visit(url)
    
    for i in range(12):
        html = browser.html
        soup = bs(html, "html.parser")
        
        browser.find_by_css('div[class="callout-media"]')[i].click()
        
        html = browser.html
        soup = bs(html, "html.parser")
        
        cat_breed = soup.find("dt", class_="statsDef-content-list-hd").get_text().replace('\n', '').strip(" ")
        weight = soup.find_all("div", class_="statsDef-content-list-item-value")[0].get_text().replace('\n', '').strip(" ")
        colors = soup.find_all("div", class_="statsDef-content-list-item-value")[2].get_text().replace('\n', '').strip(" ").replace(" ","").split(",")
        try:
            life_ex = soup.find_all("p")[3].get_text()
        except:
            print(f"No Life expectany for {cat_breed}")
            life_ex = "NA"
        
        browser.back()
        
        cat_purina_list_of_dicts.append({"breed_name" : cat_breed,
                                        "weight" : weight,
                                        "color_options" : colors,
                                        "life_expectancy" : life_ex,        
                                       })
browser.quit()
```
