```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pandas as pd
import re
import selenium
from pandas import DataFrame
```

```python
def init_browser():
    # @NOTE: Replace the path with your actual path to the chromedriver
    executable_path = {"executable_path": ChromeDriverManager().install()}
    return Browser("chrome", **executable_path, headless=False)
```

```python
browser = init_browser()
url = "https://cfa.org/breeds/"
browser.visit(url)

#THERE ARE 7 PAGES OF CAT BREEDS
#https://www.petfinder.com/cat-breeds/?page=7 - 50 Breeds Total
```

```python
html = browser.html
soup = bs(html, "html.parser")
```

```python
catfancy_breed_list_of_dicts = []

j = 35

for i in range(29,74):
    html = browser.html
    soup = bs(html, "html.parser")

    cat_breed = soup.find_all("h5", class_="elementor-heading-title elementor-size-default")[i].get_text()
    
    attributes = soup.find_all("div", class_="elementor-text-editor elementor-clearfix")[j].get_text()
    coat_length = attributes.split(":")[1].replace("Characteristics","").lstrip()
    characteristics = attributes.split(":")[2].replace("Personality","").lstrip()
    personality = attributes.split(":")[3].lstrip()

    catfancy_breed_list_of_dicts.append({"breed_name" : cat_breed,
                                        "coat_length" : coat_length,
                                        "characteristics" : characteristics,
                                        "personality" : personality,
                                       })
    j = j + 1
    
browser.quit()
```
