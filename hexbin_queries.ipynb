```python
data_df.head()
```

```python
data_df = data_df.drop(columns=['id', 'organization_animal_id','animal_type'])
```

```python
data_df
```

```python
types = data_df["type"].value_counts()
types
```

```python
data_df.drop(data_df.loc[data_df['type']=="Rabbit"].index, inplace=True)
data_df.drop(data_df.loc[data_df['type']=="Small & Furry"].index, inplace=True)
data_df.drop(data_df.loc[data_df['type']=="Bird"].index, inplace=True)
data_df.drop(data_df.loc[data_df['type']=="Scales, Fins & Other"].index, inplace=True)
data_df.drop(data_df.loc[data_df['type']=="Horse"].index, inplace=True)
data_df.drop(data_df.loc[data_df['type']=="Barnyard"].index, inplace=True)
```

```python
types = data_df["type"].value_counts()
types
```

```python
organization = data_df["organization_id"].value_counts()
organization
```

```python
orgs_df = organization.to_frame()
orgs_df
```

```python
orgs_df = orgs_df.rename_axis("organization_id")
orgs_df = orgs_df.rename(columns={"organization_id": "# of adoptable pets"})

orgs_df = orgs_df.sort_values(by = 'organization_id')

orgs_df
```

```python
mode = data_df.groupby('organization_id').age.apply(lambda x: x.mode())
print(mode)
```

```python
mode_df = mode.to_frame()
mode_df
```

```python
mode_df.drop(1, level=1, axis=0,inplace=True)
mode_df.drop(2, level=1, axis=0,inplace=True)
mode_df
```

```python
latlng_df = data_df.drop(columns=['type', 'age'])
latlng_df
```

```python
latlng_df = latlng_df.set_index('organization_id')
latlng_df
```

```python
latlng_df = latlng_df.drop_duplicates()
latlng_df
```

```python
dfs = [orgs_df, mode_df, latlng_df]
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on='organization_id'), dfs)
```

```python
df_final
```

```python
df_final.to_csv('static/data/hexbin_data.csv',index=True)
```
